```python
#importing neccesary libraries
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
#loading the data
file_path = ('D:\project\house_price_prediction\data\\train.csv')
house_data = pd.read_csv(file_path)
```

```python
#let us select target and features
Y = house_data.SalePrice

# Selecting data corresponding to features in feature_names
feature_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = house_data[feature_columns]
```

```python
#reviewing the data
X.describe()

X.head()
```

```python
#fitting the model
data_model = DecisionTreeRegressor(random_state=1)

data_model.fit(X,Y)
```

```python
predictions = data_model.predict(X)
print(predictions)
```

```python
print("First in-sample predictions:", data_model.predict(X.head()))
print("Actual target values for those homes:", Y.head().tolist())
```

```python
train_X, val_X, train_Y, val_Y = train_test_split(X,Y,random_state=1)
```

```python
val_predictions = data_model.predict(val_X)
print(val_predictions)
```

```python
#calculating MAE 
from sklearn.metrics import mean_absolute_error
val_mae = mean_absolute_error(val_Y, val_predictions)

print(val_mae)
```
